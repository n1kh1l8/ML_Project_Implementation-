# Logistic Map

# Consider the data generated by a Logistic map
$X_{n+1}$ = λ$X_n$(1 − $X_n$)

with λ = 4

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 4
N = 100
x = 0.2 + np.zeros(N)


for n in range(N - 1):
    x[n + 1] = r*x[n]*(1 - x[n])

plt.xlabel('n')
plt.ylabel('X_n')
plt.plot(x, 's-')
plt.show()
```

```python
for i in range(N - 1):
    plt.plot(x[i], x[i + 1], color='blue',marker='o')

plt.ylabel('X_n+1')
plt.xlabel('X_n')
plt.show()
```

```python
m = 5
n = 3
E = []
print(np.shape(E))
for i in range(n):
    col_vec = []
    for j in range(i, i + m):
        col_vec.append(x[j])  
    E.append(col_vec)

E = np.transpose(E)
# print(x)
print(E)
```

```python
#To get singular values of E
u, sigma, vt = np.linalg.svd(E)
np.diag(sigma)
```

```python
E_transposed = np.transpose(E)
E_transposed = np.append(E_transposed, [E_transposed[0]**2], axis = 0)
F = np.transpose(E_transposed)

print(F)
```

```python
u, sigma, vt = np.linalg.svd(F)
np.diag(sigma)
```

```python
print(vt)
```

```python
print(vt.transpose())
```

Substituting the corresponding values from the fourth column of the above matrix V in the equation $V_{1,4}(X_n)$ + $V_{2,4}(X_{n+1})$ + $V_{3,4}(X_{n+2})$ + $V_{4,4}(X_{n}^2)$ = 0, we get the equation\
$−0.696311X_n + 0.174078X_{n+1} + 0X_{n+2} + 0.696311X_{n}^2 = 0$.\
This simplifies to $4X_n − X_{n+1} − 4X^2_n = 0$ ultimately giving the equation as :\
$4X_n(1 − X_n) = X_{n+1}$ \
This gives the value of λ as 4 which is consistent with the value initially used to generate the data.

# Henon Map


# Consider the data generated by a Henon map
$X_{n+1}$ = c − a$(X_n)^2$ + $Y_n$,

$Y_{n+1}$ = $bX_n$.

with a = 1.4, b = 0.3, c = 1

```python
N = 100
X = 0.2 + np.zeros(N)
Y = 0.2 + np.zeros(N)


for n in range(N - 1):
    X[n + 1] = 1 - 1.4*X[n]*X[n] + Y[n]
    Y[n + 1] = 0.3*X[n]

plt.xlabel('n')
plt.ylabel('X_n and Y_n')
plt.plot(X, 's-', label = 'X_n')
plt.plot(Y, 's-', label = 'Y_n')
plt.legend()
plt.show()
```

```python
for i in range(N - 2):
    plt.plot(X[i], X[i + 2], color='blue',marker='o')
    plt.plot(Y[i], Y[i + 2], color='red',marker='o')

plt.ylabel('X_n+1 and Y_n+1')
plt.xlabel('X_n and Y_n')
plt.show()
```

```python
m = 7
n = 7
E1 = []

col_vec = []
for j in range(m):
    col_vec.append(1)  
E1.append(col_vec)
col_vec = []
for j in range(m):
    col_vec.append(X[j+2])  
E1.append(col_vec)
col_vec = []
for j in range(m):
    col_vec.append(X[j+1])  
E1.append(col_vec)
col_vec = []
for j in range(m):
    col_vec.append(X[j])  
E1.append(col_vec)
col_vec = []
for j in range(m):
    col_vec.append(X[j]*X[j])  
E1.append(col_vec)
col_vec = []
for j in range(m):
    col_vec.append(X[j+1]*X[j+1])  
E1.append(col_vec)
col_vec = []
for j in range(m):
    col_vec.append(X[j]*X[j+1])  
E1.append(col_vec)

E1 = np.transpose(E1)
print(E1)
```

```python
#To get singular values of E1
u, sigma, vt = np.linalg.svd(E1)
sigma
```

```python
print(vt)
```

```python
print(vt.transpose())
```

Substituting the corresponding values from the last column of the above matrix V we get the equation\
$0 = 0.496904 − 0.6956656X^2_{n+1} + 0.1490712X_n − 0.496904X_{n+2}$ as the remaining components of the V matrix are too small to be considered.\
This simplifies to $0 = 1 − 1.4X^2_{n+1} + 0.3X_n − X_{n+2}$ ultimately giving the equation as :\
$X_{n+2} = 1 − 1.4X^2_{n+1} + 0.3X_n$ \
This gives the values of a, b and c as 1.4, 0.3 and 1 respectively which is consistent with the values initially used to generate the data.

### scaled spectrum of singular values by standard SVD on the data (n2σn) and its surrogates ($n^2σ_{S1n}$ ) and  ($n^2σ_{S2n}$).

Surrogates are generated from the Fourier
Transform of the data by randomizing the phases.

```python
import numpy
def generate_fft(x):
    ts_fourier  = numpy.fft.rfft(x)
    length = len(x)
    random_phases = numpy.exp(numpy.random.uniform(0,numpy.pi,length//2+1)*1.0j)
    ts_fourier_new = ts_fourier*random_phases
    y = numpy.fft.irfft(ts_fourier_new)
    return y
```

```python
def get_sigma(data, m, e_dim, f_mat):
    F21 = []
    for i in range(e_dim):
        col_vec = []
        for j in range(i, i + m):
            col_vec.append(data[j])  
        F21.append(col_vec)

    for i in range(f_mat):
        col_vec = []
        for j in range(m):
            col_vec.append(F21[i][j]*F21[i][j])  
        F21.append(col_vec)
    F21 = np.transpose(F21)
    # print(F21)

    uf21, sigmaf21, vtf21 = np.linalg.svd(F21)
    return sigmaf21
```

```python
def plot_graphs(sigma1, sigma2, sigma3, pattern):
    x1 = []
    p1 = []
    q1 = []
    r1 = []
    for i in range(20):
        x1.append(i+1)
        p1.append((i+1)*(i+1)*sigma1[i])
        q1.append((i+1)*(i+1)*sigma2[i])
        r1.append((i+1)*(i+1)*sigma3[i])
    plt.scatter(x1, p1)
    plt.plot(x1, p1, label = f'$(n^2)${pattern}')
    plt.scatter(x1, q1)
    plt.plot(x1, q1, label = '$(n^2)$S1')

    plt.scatter(x1, r1)
    plt.plot(x1, r1, label = '$(n^2)$S2')

    plt.xlabel('n')
    # plt.ylabel('$(n^2)$σ and $(n^2)$ρ')
    plt.legend()
    plt.title('SVD and its surrogates')
    plt.show()
```

```python
y = generate_fft(x)
y2 = generate_fft(x)
```

```python
#For standard SVD
sigma1 = get_sigma(x, 21, 21, 0)
sigma2 = get_sigma(y, 21, 21, 0)
sigma3 = get_sigma(y2, 21, 21, 0)
```

```python
plot_graphs(sigma1, sigma2, sigma3, "σ-linear")
```

```python
sigma1 = get_sigma(x, 21, 11, 10)
sigma2 = get_sigma(y, 21, 11, 10)
sigma3 = get_sigma(y2, 21, 11, 10)
```

```python
#For nonlinear SVD
plot_graphs(sigma1, sigma2, sigma3, "ρ-nonlinear")
```

It is clear that the selection of nonlinear columns has worked since the nonlinear SVD has given
identically zero singular values as shown

Now, adding Gaussian and uniform noise to the data.


Note -- According to the paper,  PSNR above 28 dB for gaussian noise and 33 dB for uniform noise ensures $W_{p, p} / W_{1,1}$ goes higher than $10^{-6}$

```python
#Adding Gaussian noise to the data
mu=0.0
std = 0.25 * np.std(x) # for 25% Gaussian noise
def gaussian_noise(x,mu,std):
    noise = np.random.normal(mu, std, size = x.shape)
    x_noisy = x + noise
    return x_noisy 

x_noisy = gaussian_noise(x, mu, std)
y_noisy = generate_fft(x_noisy)
y2_noisy = generate_fft(x_noisy)
```

```python
from math import log10, sqrt
from sklearn.metrics import mean_squared_error


def PSNR(original, compressed):
    rms = mean_squared_error(original, compressed, squared=False)
    rms = np.mean((original - compressed) ** 2)
    if(rms == 0):
        return 100
    max_value = max(max(compressed), max(x))
    max_pixel = max_value
    psnr = 20 * log10(max_pixel / rms)
    return psnr
```

#Application to Mathematical Modeling

```python
import numpy as np
from scipy import integrate
from matplotlib import pyplot as plt
import time
 
k = 2.895    
c = 0.237  
def flow_deriv(x_y,tspan):
    x, y = x_y
    a = y
    b = c*y*(1-x**2) - k*x
    return[a,b]
 
px1 = np.random.rand(1)
w1 = 0
 
x_y = [px1, w1]
 
t = np.linspace(0, 2000, 40000)
x_t = integrate.odeint(flow_deriv, x_y, t)
x0 = x_t[39999,0:2]   
 
tspan = np.linspace(1,20000,400000)
x_t = integrate.odeint(flow_deriv, x0, tspan)
siztmp = np.shape(x_t)
siz = siztmp[0]
 
y1 = x_t[:,0]
y2 = x_t[:,1]

print("Phase Space of the Vander Pol Equation") 
plt.figure(2)
lines = plt.plot(y1[1:20000],y2[1:20000],'ko',ms=1)
plt.setp(lines, linewidth=0.5)
plt.xlabel("Xs,Zn")
plt.ylabel("Vs/Vn")
plt.show() 
```

```python
t_a = np.linspace(1,100,10000) # a series of size 10,000 is generated by team A from the differential equations given
x_a = integrate.odeint(flow_deriv, x0, t_a)
t_b = np.linspace(2,20,1000)
x_b = integrate.odeint(flow_deriv, x0, t_b) # 1000 values from the series

y1b = x_b[:,0]
y2b = x_b[:,1]
     
plt.figure(2)
lines = plt.plot(y1b[1:1000],y2b[1:1000],'ko',ms=1)
plt.setp(lines, linewidth=0.5)
plt.show()  # graph of phase space of the shorter series 

x_bbb = np.zeros(1000)  # calculate the first and second derivatives of the time series
y1_b = y1b
for i in range(0,999):
  x_bbb[i] = y1_b[i+1]
y2_b = (x_bbb - y1_b)/0.01
for i in range(0,998):
  x_bbb[i] = y2_b[i+1]
y3_b = (x_bbb - y2_b)/0.01
```

```python
F_b = [] # to generate and store the embedding matrix

col_vec = []
for j in range(0, 998):
    col_vec.append(y3_b[j])  
F_b.append(col_vec)

col_vec = []
for j in range(0, 998):
    col_vec.append(y1_b[j])  
F_b.append(col_vec)

col_vec = []
for j in range(0, 998):
    col_vec.append(y2_b[j])  
F_b.append(col_vec)

col_vec = []
for j in range(0, 998):
    col_vec.append(y1_b[j]**2)  
F_b.append(col_vec)

col_vec = []
for j in range(0, 998):
    col_vec.append(y2_b[j]**2)  
F_b.append(col_vec)

col_vec = []
for j in range(0, 998):
    col_vec.append(y1_b[j]*y2_b[j])  
F_b.append(col_vec)

F_b = np.transpose(F_b)

print(F_b)
```

```python
u, sigma, vt = np.linalg.svd(F_b)
sigma # calculate singular values of the matrix
```

```python
print(vt.transpose())
```

The linear relationship recovered gives the Van der Pol equation with parameter values k = 2.895 and c = 0.237 which were the same as used to generate the data.
